```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.read_csv("polynomial.csv", sep = ";")
```

```python
polynomial_regression = PolynomialFeatures(degree = 2)
```

```python
#Veri setine bir göz atmak için df yazdırılıyor:
df
```

```python
#Veriler, deneyim yılları ve maaş arasındaki ilişkiyi gösteren bir scatter plot (nokta grafiği) ile görselleştiriliyor:
plt.scatter(df['deneyim'],df['maas'])
plt.xlabel('deneyim(yil)')
plt.ylabel('maas')
plt.savefig('1.png', dpi=300)
plt.show()
```

```python
#Daha sonra, doğrusal bir regresyon modeli (LinearRegression) kullanılarak verilerin lineer regresyonunu alıyoruz:
reg = LinearRegression()
reg.fit(df[['deneyim']],df['maas'])
#Bu model, verilerin cunu buluyor ve bu verilere uygun bir doğru çiziyor.
```

```python
#Sonuçları görselleştirmek için lineer regresyon tahmin doğrusu çiziliyor:
plt.xlabel('deneyim(yil)')
plt.ylabel('maas') #Grafikte x ekseni için 'deneyim(yıl)' ve y ekseni için 'maaş' etiketlerini belirtir.
plt.scatter(df['deneyim'],df['maas']) #Veri setindeki 'deneyim' ve 'maas' değerlerini içeren bir scatter plot (nokta grafiği) oluşturur. 
xekseni = df['deneyim']
yekseni = reg.predict(df[['deneyim']]) # Lineer regresyon modelini kullanarak 'deneyim' değerlerine karşılık gelen 'maas' tahminlerini yapar. Bu tahminler yekseni adlı bir değişkene atanır.
plt.plot(xekseni, yekseni, color="green",label="linear regression") #Lineer regresyon modeli tarafından yapılan 'maas' tahminlerini içeren doğruyu çizer. Bu doğru, 'deneyim' değerlerine karşılık gelen tahmin edilen 'maas' değerlerini gösterir. color="green" ile doğrunun rengi belirtilir ve label="linear regression" ile bu doğrunun grafiğin açıklamasında görünecek etiket belirtilir.
plt.legend() #Grafikteki açıklamayı (legend) etkinleştirir. Bu, her bir grafiğin neyi temsil ettiğini gösteren küçük bir açıklama kutusu oluşturur.
plt.show()
#Bu adımda, lineer regresyon modeli tarafından tahmin edilen maaşları gösteren bir doğru çizilir.
```

```python
polynomial_regression = PolynomialFeatures(degree = 2)
x_polynomial = polynomial_regression.fit_transform(df[['deneyim']])
```

```python
reg = LinearRegression()
reg.fit(x_polynomial,df['maas'])
```

```python
#Son olarak, polinomlu regresyon modeli kullanılarak maaş tahminleri yapılır ve bu tahminler görselleştirilir:
y_head = reg.predict(x_polynomial)
plt.plot(df['deneyim'],y_head,color="red", label = "polynomial regression") # df['deneyim'] (deneyim yılları) verilerine karşılık gelen y_head (polinomlu regresyonun tahminleri) verileri grafiğe çizilir
plt.legend() # metodu, grafiğin sol üst köşesinde "polynomial regression" etiketini gösterir
plt.scatter(df['deneyim'],df['maas']) #plt.scatter() ile gerçek veriler (df['deneyim'] ve df['maas']) grafiğe nokta olarak çizilir
plt.show()
```

```python
#on olarak, bir örnekleme yaparak modelin tahminlerini alabiliriz:
x_polynomial1 = polynomial_regression.fit_transform([[4.5]])
reg.predict(x_polynomial1)
```
