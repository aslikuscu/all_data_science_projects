````python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
````

````python
sonar_data = pd.read_csv('Copy of sonar data.csv', header=None)
````

````python
sonar_data.head()
````

Bu veri, **sonar sinyalleri** kullanılarak **kayaları (R)** ve **mayınları (M)** ayırt etmeyi amaçlayan bir veri setidir. Her satır, bir sonar sinyalinin çeşitli frekanslardaki ölçümlerini temsil eder ve son sütun, sinyalin bir kayadan mı yoksa mayından mı yansıdığını gösterir.

### Verideki Detaylar:
- **0-59 arası sütunlar**: Sinyalin farklı frekanslardaki yansımalarını temsil eden sayısal değerlerdir. Yani, her sütun bir frekansa ait sinyal gücünü gösterir.
- **60. sütun**: Son sütun, sinyalin yansıdığı nesne türünü belirtir:
  - `R`: Kaya (Rock)
  - `M`: Mayın (Mine)

### Gözlemler:
- **Değerlerin aralığı**: 0 ile 1 arasında değişen küçük sayılar. Bu, sinyal genliklerinin normalleştirildiğini gösteriyor.
- **Etiketler**: Verinin sınıflandırma problemi için hazır olduğu görülüyor çünkü her satırın sonunda bir etiket (R veya M) var.

Sonuç olarak, bu veri, sonar sinyallerinin hangi tür yüzeyden yansıdığını tahmin etmek için kullanılıyor ve makine öğrenmesi modelleriyle çalışmak için uygun.

- **R (Rocks)**: 97 kez gözlemlenmiş.

````python
X = sonar_data.drop(columns=60, axis=1)
Y = sonar_data[60]
````

````python
print(X)
print(Y)
````

--

### **Adımların Detaylı Açıklaması:**

1. **Verileri Hazırlama:**
   - **X (Girdiler)**: 0-59 sütunları, yani ses sinyalinin frekans genlikleri. Model bu verilerden öğrenir.
   - **Y (Çıktı/Etiket)**: 60. sütun, yani gözlemin **Mine (M)** veya **Rock (R)** olduğunu belirten sınıf etiketi.

2. **Modeli Eğitme:**
   - Model, **X'teki sinyal verilerini** kullanarak **Y'nin sınıfını** tahmin etmeyi öğrenir.
   - Bu süreçte model, **girdiler (X)** ile **etiketler (Y)** arasındaki ilişkileri keşfetmeye çalışır.

3. **Tahmin Yapma:**
   - Eğitim tamamlandıktan sonra model, yeni verilere (yeni X değerlerine) bakarak bir tahmin yapar. Bu tahmin, gözlemin **Mine** mı yoksa **Rock** mu olduğunu söyler.

4. **Tahmin Sonucunu Değerlendirme:**
   - Modelin yaptığı tahminler, gerçek etiketlerle (**Y'nin gerçek değerleriyle**) karşılaştırılır.
   - **Hedef:** Modelin tahminlerinin doğruluğunu ölçmek ve hataları azaltmak.

---

### **Bir Örnekle Açıklama:**

- **Eğitim Verisi (X):**  
  - Girdiler: `[0.0200, 0.0371, 0.0428, ..., 0.0032]` → Ses sinyali verileri.
  
- **Etiket (Y):**  
  - Gerçek sınıf: `'R'` (Rock).

- **Modelin Tahmini:**  
  - Model bu girdileri kullanarak `'R'` (Rock) olduğunu tahmin eder.

- **Değerlendirme:**  
  - Modelin tahmini gerçek değerle (Y) karşılaştırılır. Eğer doğruysa, başarı; yanlışsa hata olarak kaydedilir.

---

### **Sonuç:**
- **Amaç:** Modeli eğiterek, sinyal verilerine (X) bakarak doğru sınıfı (Y) tahmin edebilmesini sağlamaktır. Tahmin edilen Y ile gerçek Y'yi karşılaştırarak modelin performansını ölçeriz.

Bu döngüyü iyileştirmek, modelin doğruluğunu artırır! 🚀

````python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify=Y, random_state=1)
````

````python
print(X.shape, X_train.shape, X_test.shape)
````

````python
print(X_train)
print(Y_train)
````

````python
model = LogisticRegression()
````

````python
model.fit(X_train, Y_train)
````

````python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train) 
````

````python
print('Accuracy on training data : ', training_data_accuracy)
````

#### **Eğitim (Training):**
```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```
- Bu adımda, **Logistik Regresyon modelini** oluşturuyorsunuz ve modelin eğitimini **X_train** (özellikler) ve **Y_train** (etiketler) verisiyle yapıyorsunuz.
- **model.fit()** fonksiyonu, modelin **özellikler ile etiketler arasındaki ilişkiyi öğrenmesini** sağlar. Bu aşama, modelin sınıfları (Mines ve Rocks) daha iyi ayırt edebilmesi için bir model oluşturur.

#### **Modelin Performansını Değerlendirme:**
```python
model.score(X_test, Y_test)
```
- **score()** fonksiyonu, modelin **test verisi** üzerinde ne kadar başarılı olduğunu gösterir. Bu, modelin genel **doğruluk oranını** ölçer.
  - **X_test** ve **Y_test**, **test veri kümesi** üzerindeki özellikler ve gerçek etiketlerdir.
  - Test doğruluğu: **0.8333 (yaklaşık %83)**. Bu, modelin test veri kümesindeki örnekleri doğru tahmin etme oranıdır.

```python
model.score(X_train, Y_train)
```
- Eğitim verisi üzerindeki doğruluk: **0.8072 (yaklaşık %81)**. Bu, modelin **eğitim verisi** üzerindeki doğruluğudur.
  
- Bu sonuçlar şunları gösteriyor:
  - Eğitim verisi üzerinde doğruluk oranı biraz daha düşükken, test verisi üzerinde doğruluk oranı **%83** civarında. Bu, modelin **genelleme** yeteneğinin iyi olduğunu ve **overfitting (aşırı öğrenme)** yapmadığını gösteriyor. Yani, model **eğitim verisinde iyi performans gösterdiği kadar test verisinde de başarılı**.

---

### **Sonuçların Yorumlanması:**

- **Modelin Test Performansı (0.8333)**: Test verisi üzerindeki doğruluk oranı oldukça yüksek, yani model **yeni ve görünmeyen verilerle iyi sonuçlar** veriyor. Bu, modelin **genelleme kapasitesinin iyi olduğunu** gösterir.
  
- **Modelin Eğitim Performansı (0.8072)**: Eğitim verisi üzerindeki doğruluk oranı biraz daha düşük. Bu, modelin eğitim verisini öğrenip, daha fazla **özelliklerin etkisini** anlamaya başladığını, ancak aynı zamanda test verisinde **daha iyi sonuçlar** verdiğini gösterir.

### **Logistik Regresyonun Seçilmesinin Nedeni:**
- **Verinin doğrusal doğası**: Sonar verisi, **ikili sınıflandırma** ve **doğrusal sınırlarla** iyi ayırt edilebilecek veriler içeriyorsa, logistik regresyon iyi sonuçlar verebilir.
---

### Durum:
- **60 tane sütun (özellik)** var.
- Eğer 45 tanesi **'R' (Rock)** sınıfına aitse, **kalan 15 tanesi ise 'M' (Mine)** sınıfına aitse, **bu durum şu şekilde değerlendirilir**:

### Modelin Tahmin Yöntemi:
1. **Logistik regresyon** modeli her bir sütun için olasılık değerleri hesaplar. Bu, her sütunun **'R'** veya **'M'** sınıfına ait olma olasılıklarıdır.
2. Modelin genel **tahmininin ne olacağı**, her sütun için hesaplanan **olay olasılıklarının bir araya gelmesiyle belirlenir**. Modelin verdiği olasılık değeri **0 ile 1 arasında bir sayıdır**. 
   - Eğer olasılık **0.5'in üstünde** ise, model o örneği **'M' (Mine)** sınıfına atar.
   - Eğer olasılık **0.5'in altında** ise, model o örneği **'R' (Rock)** sınıfına atar.

### Durumun Gerçek Yorumlanması:
- Eğer **60 sütundan 45'i 'R' (Rock)** sınıfına aitse, bu durumda modelin büyük olasılıkla **'R' (Rock)** sınıfına ait olduğunu tahmin etmesi beklenir.
- Çünkü çoğunluk **'R' (Rock)** sınıfından oluşuyor ve model bu durumu daha olası olarak değerlendirir. Modelin tahmin ettiği olasılık, bu durumda **'R' (Rock)** sınıfına daha yakın olacaktır.


 Burada aslında **"tahminleme"** yapmıyoruz, çünkü elimizdeki veriler zaten etiketlenmiş ve her bir örneğin **"Rock" (R) veya "Mine" (M)** olduğu belli. Ancak, yapılan işlem şudur:

### 1. **Model Eğitimi (Training)**:
- Elimizde **etiketlenmiş veriler** var. Bu verilerde her bir satırda 60 özellik (sinyal verileri) ve 1 etiket (R veya M) bulunuyor.
- **Logistik regresyon** modeli bu verileri kullanarak eğitildi. Model, sinyal verilerinin (özelliklerin) **"Rock" (R)** veya **"Mine" (M)** sınıfına ait olup olmadığını öğrenmeye çalıştı.

### 2. **Tahmin (Prediction)**:
- Elimizde bir **yeni örnek veri** (yani **input_data**) olduğunda, model bu veriyi kullanarak bir tahminde bulunur. Model, bu örneğin **"Rock" (R)** mi yoksa **"Mine" (M)** mi olduğunu tahmin eder.
  
- Yani bu durumda model, daha önce **eğitilen verilerden öğrendiği ilişkiler** doğrultusunda bu yeni veriyi **'R' (Rock)** ya da **'M' (Mine)** sınıfına yerleştirir.


### Özet:
- **Tahminleme** yaptık çünkü model yeni veriyi kullanarak **'R' (Rock)** ya da **'M' (Mine)** sınıflandırması yaptı.
- Bu sınıflandırma aslında **doğru etiketleme** yapıldığının kontrolüydü, çünkü zaten verilerin etiketleri biliniyordu (sonuçları doğrulamak için).
